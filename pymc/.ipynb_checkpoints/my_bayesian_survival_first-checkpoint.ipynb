```python
%matplotlib inline
import numpy as np
from IPython.core.pylabtools import figsize
import matplotlib.pyplot as plt
import scipy.stats as stats
import pymc as pm
import random

from lifelines.plotting import plot_lifetimes
```

```python
# create some data and plot it

N = 1000
obs_time = 10
random.seed(12345)

lifetime = pm.rweibull(2, 5, size=N)
birth = pm.runiform(0, 10, N)
death = birth + lifetime

censor = death >= obs_time
lifetime_ = np.ma.masked_array(lifetime, censor)
lifetime_.set_fill_value(obs_time)
```

```python
# plot it!

figsize(12.5, 10)

plt.xlim(0, 25)
plt.vlines(obs_time, 0, N, lw=2, linestyles="--")
plt.xlabel('time')
plt.title('Births and deaths, at $t=%d$' % obs_time)
plot_lifetimes(lifetime, birthtimes=birth,
              event_observed = ~ censor)
```

```python
# the model

alpha = pm.Uniform("alpha", 0, 20)
beta = pm.Uniform("beta", 0, 20)
obs = pm.Weibull("obs", alpha, beta, value = lifetime_, observed = True)

@pm.potential
def censor_factor(obs=obs):
    if np.any((obs+birth < obs_time)[censor]):
        return -100000
    else:
        return 0
    
mcmc = pm.MCMC([alpha, beta, obs, censor_factor])
mcmc.sample(100000, 30000)

alpha_t = mcmc.trace("alpha")[:]
beta_t = mcmc.trace("beta")[:]
```

```python
medians = beta_t * np.log(2) ** (1/alpha_t)
print (medians.mean(), np.median(lifetime_), np.median(lifetime))

figsize(12.5, 6)

# histogram of median samples:
plt.title(r"Posterior distributions of median")
plt.hist(medians, histtype='stepfilled', bins=35, alpha=0.85)
plt.vlines(np.median(lifetime), 0, 1000, lw=2, linestyles="--")
plt.legend()
plt.show()

figsize(12.5, 6)

# histogram of the samples:
plt.title(r"Posterior distributions of $\alpha, \beta$")
plt.hist(beta_t, histtype='stepfilled', bins=35, alpha=0.85,
         label=r"posterior of $\beta$", color="#7A68A6", normed=True)
plt.hist(alpha_t, histtype='stepfilled', bins=35, alpha=0.85,
         label=r"posterior of $\alpha$", color="#A60628", normed=True)
plt.vlines(2, 0, 1.2, lw=2, linestyles="--")
plt.vlines(5, 0, 1.2, lw=2, linestyles="--")


plt.legend()

from pymc.Matplot import plot as mcplot
#mcplot(mcmc)
```
