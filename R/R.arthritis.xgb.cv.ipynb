```r
require(xgboost)
require(vcd)
require(Matrix)
require(data.table)

data("Arthritis")
head(Arthritis)
```

```r
y <- ifelse(Arthritis$Improved == "Marked", 1, 0)

df <- Arthritis %>%
    mutate(AgeDiscret = round(Age/10,0)) %>%
    select(-ID, -Age)

X = sparse.model.matrix(Improved~.-1, data = df)

head(X)
```

```r
set.seed(1234)
cv <- xgb.cv(params = list(max.depth = 3,
                           eta = 0.2, 
                           nthread = 2, 
                           objective = "binary:logistic"),
             nrounds = 5,
             data = X,
             label = y,
             metrics = list("error", "auc"),
             early.stop.round = 3,
             maximize = FALSE,
             nfold = 7)
```
