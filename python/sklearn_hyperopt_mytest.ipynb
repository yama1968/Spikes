```python
import sklearn.datasets
import numpy as np
import random
from math import log

data = sklearn.datasets.load_digits()
X = data["data"]
y = data["target"]

from sklearn.svm import SVC
from sklearn.cross_validation import StratifiedKFold
from sklearn.cross_validation import cross_val_score

# define an objective function
def objective(args):
    
    print args,
    
    clf = SVC(**args)
    this_scores = cross_val_score(clf, X, y, scoring="accuracy")
    this_score = this_scores.mean()

    print("%.4f (%.4f)" % (this_score, this_scores.std()))
    
    return -this_score
              

# define a search space
from hyperopt import fmin, tpe, hp, STATUS_OK, Trials

space ={
    'C':      hp.loguniform('C', log(1e-9), log(1e9)),
    'gamma':  hp.loguniform('gamma', log(1e-9), log(1e9))
}


trials = Trials()

%time best = fmin(objective, space, algo=tpe.suggest, max_evals=100, trials=trials)

print best
```
