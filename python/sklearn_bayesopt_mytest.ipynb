```python
import sklearn.datasets
import numpy as np
import random
from math import log,exp

from __future__ import print_function
from __future__ import division

from sklearn.cross_validation import cross_val_score
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.svm import SVC

from bayes_opt import BayesianOptimization


data = sklearn.datasets.load_digits()
X = data["data"]
y = data["target"]


# define an objective function
def objective(**kwargs):
    
    print (kwargs)
    
    fargs = {'C': exp(kwargs['C']), 'gamma': exp(kwargs['gamma'])}
    
    print (fargs)
    
    clf = SVC(**fargs)
    this_scores = cross_val_score(clf, X, y, scoring="accuracy")
    this_score = this_scores.mean()

    print("%.4f (%.4f)" % (this_score, this_scores.std()))
    
    return this_score
              

space ={
    'C':      (log(1e-9), log(1e9)),
    'gamma':  (log(1e-9), log(1e1))
}


svcBO = BayesianOptimization(objective, space)
svcBO.explore({'C': [-3, -2, 1], 'gamma': [-3, -2, -1]})

%time svcBO.maximize(n_iter=120)

print (svcBO.res["max"])
```
