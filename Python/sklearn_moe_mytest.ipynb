```python
import sklearn.datasets
import numpy as np
import random
import moe

data = sklearn.datasets.load_digits()
X = data["data"]
y = data["target"]

from sklearn.svm import SVC
from sklearn.cross_validation import cross_val_score

from moe.easy_interface.experiment import Experiment
from moe.easy_interface.simple_endpoint import gp_next_points
from moe.optimal_learning.python.data_containers import SamplePoint


def p2p(point):
    return {'C': 10**point[0], 'gamma': 10**point[1], 'kernel': 'rbf'}

def optimize(expe, n = 100):

    best_score = 0
    
    for i in range(n):
        next_point = gp_next_points(expe)[0]

        print p2p(next_point), " -> ",

        clf = SVC(**p2p(next_point))

        this_scores = cross_val_score(clf, X, y, scoring="accuracy")
        this_score = this_scores.mean()

        print("%.4f (%.4f)" % (this_score, this_scores.std()))

        if this_score > best_score:
            best_score = this_score
            best_expe  = expe
            best_point = next_point

        expe.historical_data.append_sample_points(
            [SamplePoint(next_point, -this_score)])

    return best_score, best_point


expe = Experiment([[-9, 9], [-9, 0]])

%time best_score, best_point = optimize(expe)
        
print ("best is %.4f, with %s" % (best_score, str(p2p(best_point))))
```
